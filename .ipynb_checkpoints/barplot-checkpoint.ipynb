```python
import seaborn as sns
import matplotlib.pylab as plt
import numpy

# Load data set
kashti = sns.load_dataset("titanic")
kashti
# draw a line plot

# Create a diverging palette with a darker center
# dark_center_div = sns.diverging_palette(100, 275, s=80, l=55, n=9, center="light")


sns.barplot(x="sex", y="alone", hue="who", data=kashti, order=["female", "male"], palette="mako_r", errorbar=None)


plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load dataset
kashti = sns.load_dataset("titanic")

# Set plot background color
plt.figure(facecolor='lightgrey')  # Set figure face color for the outer background
ax = sns.barplot(x="fare", y="class", hue="sex", data=kashti, saturation=0.2, err_kws={"color": "red"})

# Set inner plot (axes) face color
ax.set_facecolor("lightgreen")  # Customize as desired

plt.show()
```
